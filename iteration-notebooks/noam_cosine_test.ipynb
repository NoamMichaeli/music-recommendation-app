```python
# Import necessary libraries
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Define the user-item ratings matrix
data = {
    "Item 1": [0, 4, 0, 3, 4],
    "Item 2": [3, 0, 0, 0, 3],
    "Item 3": [0, 0, 3, 4, 0],
    "Item 4": [3, 2, 0, 0, 4],
    "Item 5": [4, 0, 0, 3, 4]
}
```

```python
# User indices
users = ["User 1", "User 2", "User 3", "User 4", "User 5"]
```

```python
# Create a Pandas DataFrame
df = pd.DataFrame(data, index=users)
```

```python
# Display the user-item matrix
print("User-Item Ratings Matrix:")
display(df)
```

```python

# Compute cosine similarity between users
cosine_sim = cosine_similarity(df)
```

```python
# Create a DataFrame for the similarity matrix
cosine_sim_df = pd.DataFrame(cosine_sim, index=users, columns=users)
```

```python
# Display the User-User Cosine Similarity Matrix
print("\nUser-User Cosine Similarity Matrix:")
display(cosine_sim_df)
```
